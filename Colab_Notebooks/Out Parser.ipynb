```python
from matplotlib import pyplot as plt
```

```python
file_direct = './gridSearch/'
def getModelMeans(filename):
    mean_list = list()
    with open(file_direct + filename, "r") as jsonfile:
        line = jsonfile.readline()
        while line:
            if line[0] != 'm':
                line = jsonfile.readline()
                continue
            line = line.split(" ")
            mean_list.append(float(line[1]))
            line = jsonfile.readline()
        return mean_list
```

```python
file_list = ['AfterSuperBowl.txt', 'BeforeSuperBowl.txt', 'DuringSuperBowl.txt', 'GradientBoostRegressorP8.txt', 'RandomForestRegressorP8.txt']
error_list = list()
for filename in file_list:
    mean_list = getModelMeans(filename)
    print(filename, "Optimal parameter Cross Validation Error: ", max(mean_list))
    error_list.append(mean_list)
```

```python
def plot_array(arr, lbl):
    krange=range(len(arr))
    plt.plot(krange, arr, label=lbl)
    plt.legend(loc='best')
    plt.title("Cross Validation Error")
    plt.show()
```

```python
for means, filename in zip(error_list, file_list):
    plot_array(means, filename.split(".")[0])
```
